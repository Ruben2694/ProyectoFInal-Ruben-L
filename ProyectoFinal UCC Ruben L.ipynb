```python
bins = np.linspace(min(EnergiasRen["1990"]), max(EnergiasRen["1990"]), 4)
bins
```

```python
nombre = ["pequeño", "mediano", "grande"]
EnergiasRen["1990-bins"] = pd.cut(EnergiasRen["1990"], bins, labels=nombre, include_lowest=True)
```

```python
EnergiasRen.head(10)
```

```python
EnergiasRen[["1990", "1990-bins"]].head(10)
```

```python
EnergiasRen[["1990-bins"]].value_counts()
```

```python
plt.hist(EnergiasRen["1990-bins"])
plt.xlabel("Cantegorias 1990")
plt.ylabel("Cantidad paises")
plt.title("Cantidad de energia renovable usada% en total por pais")
```

```python
import matplotlib.cbook as cbook
 
fig, ax = plt.subplots()
ax.scatter(EnergiasRen['1990'], EnergiasRen['2005'], alpha=0.5)


ax.set_xlabel("1990", fontsize=15)
ax.set_ylabel("Uso de energia Renovable", fontsize=15)
ax.set_title('Correlacion 1990 vrs 2005')
 
ax.grid(True)
fig.tight_layout()
 
plt.show()
```
